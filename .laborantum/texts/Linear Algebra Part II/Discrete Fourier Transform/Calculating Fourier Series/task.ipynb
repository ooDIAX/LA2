```python
import numpy as np
import json_tricks
import scipy.integrate
```

```python
k = np.arange(100)

f1 = lambda x: 1
f2 = lambda x: np.exp(x)
f3 = lambda x: x**2 - x

def compute_fourier_coefficients(f, k_values):
    coefficients = []
    for n in k_values:
        real_part = scipy.integrate.quad(lambda x: np.real(f(x) * np.exp(-1j * 2 * np.pi * n * x)), 0, 1, epsabs=1e-12, epsrel=1e-12)[0]
        imag_part = scipy.integrate.quad(lambda x: np.imag(f(x) * np.exp(-1j * 2 * np.pi * n * x)), 0, 1, epsabs=1e-12, epsrel=1e-12)[0]
        coefficients.append(real_part + 1j * imag_part)
    return np.array(coefficients, dtype=np.complex128)
  
coefficients_f1 = compute_fourier_coefficients(f1, k)
coefficients_f2 = compute_fourier_coefficients(f2, k)
coefficients_f3 = compute_fourier_coefficients(f3, k)

answer = {
    'task1': coefficients_f1,
    'task2': coefficients_f2,
    'task3': coefficients_f3,
}

json_tricks.dump(answer, '.answer.json')
```
