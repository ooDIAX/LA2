```python
import numpy as np
import json_tricks
```

```python
def task_1(x, y):
    J = np.array([[2 * x, 1], [y, x]])
    det_J = np.linalg.det(np.array([[2 * x, -1], [y, x]]))
    return J, det_J

def task_2(r, phi):
    J = np.array([[np.sin(phi), r * np.cos(phi)],
                  [np.cos(phi), -r * np.sin(phi)]])
    return J, np.linalg.det(J)
```

```python
inputs = json_tricks.load('inputs.json')

results = {
    'task1': [],
    'task2': []
}

for args in inputs['task1']:
    print(inputs['task1'])
    results['task1'].append(task_1(**args))

for args in inputs['task2']:
    results['task2'].append(task_2(**args))

json_tricks.dump(results, '.answer.json')
```
